# OME-TIFF fix for files generated by Imaris

When we generate OME-TIFF files from Imaris, they qre incorrectly written and change the Z resolution by T.
With this notebook we can parse through the OME-TIFF files in one folder, split 3 channels and save them correctly and independently.

## Setting up the Requisites

```python
import os
import tifffile as tf
```

## Folder location
Change it to where your files are located, this is also the destination folder

```python
data = r'F:\Oct4 validated\12X_1Z'
```

## Files parsing
In this example the resolution is 0.542, 0.542 and 2 microns in XYZ, change the units according to your acquisitions.

```python
for filename in os.listdir(data):
    if filename.endswith(".tif"):
        im = tf.imread(data+'/'+filename)
        #channel0 = im[:,:,:]
        #channel1 = im[:,1,:,:]
        #channel2 = im[:,2,:,:]
        c=0
        name = filename+'_C'+str(c)+'.tif'
        dtype = im.dtype
        shape = im.shape
        
        print(shape)
        
        tf.imwrite(data+'/'+name,
                   im,
                   im.astype(dtype),
                   bigtiff=True,
                   compression='zlib',
                   resolution=(0.542, 0.542),
                   metadata={'spacing': 2,'unit': 'um','axes': 'ZYX'})
        c+=1
        
        #name = filename+'_C'+str(c)+'.tif'
        #dtype = im.dtype
        #shape = im.shape
        #tf.imwrite(data+'/'+name,
                   #channel1,
                   #im.astype(dtype),
                   #imagej=True,
                   #compression='zlib',
                   #resolution=(1./0.542, 1./0.542),
                   #metadata={'spacing': 2,'unit': 'um','axes': 'ZYX',})
        
        #c+=1
        
        #name = filename+'_C'+str(c)+'.tif'
        #dtype = im.dtype
        #shape = im.shape
        #tf.imwrite(data+'/'+name,
                   #channel2,
                   #im.astype(dtype),
                   #imagej=True,
                   #compression='zlib',
                   #resolution=(1./0.542, 1./0.542),
                   #metadata={'spacing': 2,'unit': 'um','axes': 'ZYX',})
        
        
print('all files fixed')
```
